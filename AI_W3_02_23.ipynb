```python
# Load library
import pandas as pd
# Create DataFrame
dataframe = pd.DataFrame()
# Add columns
dataframe['Name'] = ['Lemon', 'Apple','Banana']
dataframe['Price'] = [5,8,7]
dataframe['license'] = [False,True,True]
```

```python
df = dataframe
df.info()
```

```python
df.iloc[0:2]
```

```python
df.loc[(df.Price == 5)| (df.Name == 'Banana')]
```

```python
df.loc[(df.Price == 5)| (df.Name == 'Banana'), ['Name','Price']]
```
