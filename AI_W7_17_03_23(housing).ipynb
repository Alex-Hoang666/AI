```python
import pandas as pd
url='/content/Housing.csv'
data=pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlations=data.corr()
import seaborn as sns
sns.heatmap(correlations,annot=True)
sns.pairplot(data)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['area']), y=data['price'])
print('he so intercept',model.intercept_)
print('he so coef',model.coef_)
model.score(X=pd.DataFrame(data['area']),y=data['price'])
y_pred=model.predict(X=pd.DataFrame(data['area']))
data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='black')
plt.plot(data['area'],y_pred,color='blue')
```

```python
#Bc 2
model=linear_model.LinearRegression()
X=pd.DataFrame([data['area'] ,data['area']**2]).T  #chuyn hng thnh ct
model.fit(X,y=data['price'])
print('Intercept ca m hnh:',model.intercept_)
print('coef cua mo hinh:',model.coef_)
print(' chnh xc:',model.score(X,y=data['price']))
import numpy as np
range=np.arange(0,13300000,10000)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
