```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url='/content/insurance.csv'
data=pd.read_csv(url)
data.sample()
data.info()
```

```python

print(data['sex'].value_counts())
sns.countplot(x='sex',data=data)
print(data['smoker'].value_counts())
sns.countplot(x='smoker',data=data)
```

```python
print(data['smoker'].value_counts())
sns.countplot(x='smoker',data=data)
```

```python
print(data['region'].value_counts())
sns.countplot(x='region',data=data)
```

```python
corr=data.corr()
sns.heatmap(corr,cmap='Wistia',annot=True)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['age']), y=data['charges'])
print('he so intercept',model.intercept_)
print('he so coef',model.coef_)
model.score(X=pd.DataFrame(data['age']),y=data['charges'])
y_pred=model.predict(X=pd.DataFrame(data['age']))
data.plot(kind='scatter',x='age',y='charges',figsize=(9,9),color='black')
plt.plot(data['age'],y_pred,color='blue')
```

```python
#Bậc 2
model=linear_model.LinearRegression()
X=pd.DataFrame([data['age'] ,data['age']**2]).T  #chuyển hàng thành cột
model.fit(X,y=data['charges'])
print('Intercept của mô hình:',model.intercept_)
print('coef cua mo hinh:',model.coef_)
print('Độ chính xác:',model.score(X,y=data['charges']))
import numpy as np
range=np.arange(0,100,1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='age',y='charges',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```

```python
#Bậc 3
model=linear_model.LinearRegression()
X=pd.DataFrame([data['age'] ,data['age']**2,data['age']**3]).T  #chuyển hàng thành cột
model.fit(X,y=data['charges'])
print('Intercept của mô hình:',model.intercept_)
print('coef cua mo hinh:',model.coef_)
print('Độ chính xác:',model.score(X,y=data['charges']))
range=np.arange(0,100,1)
X_range=pd.DataFrame([range,range**2,range**3]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='age',y='charges',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```

```python
#bậc 10
model=linear_model.LinearRegression()
X=pd.DataFrame([data['age'] ,data['age']**2,data['age']**3,data['age']**4,data['age']**5,
              data['age']**6,data['age']**7,data['age']**8,data['age']**9,data['age']**10]).T  #chuyển hàng thành cột
model.fit(X,y=data['charges'])
print('Intercept của mô hình:',model.intercept_)
print('coef cua mo hinh:',model.coef_)
print('Độ chính xác:',model.score(X,y=data['charges']))
import numpy as np
range=np.arange(0,100,1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='age',y='charges',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
